## 3. New Architectures for Cognitive Modeling

These architectures go beyond vanilla RNNs and are increasingly used as models of cognition. See **Appendix A3** for hands-on implementations.

### State Space Models
- **BrainMamba** (CHIL 2024): S4/Mamba for brain activity encoding. Outperforms prior methods on neural decoding with linear complexity in sequence length. [arXiv](https://arxiv.org/abs/2403.01190)
- **NeuroMamba** (NeurIPS 2025): Extends Mamba architecture to neural population decoding.
- *Module connection:* SSM updates resemble Kalman filter steps — a direct link to Bayesian inference (Module 8).

### Continuous-Time Networks
- **Closed-form Continuous-time Networks (CfCs)** — Hasani, Lechner et al., *Nature Machine Intelligence* 2022. Inspired by biological neural dynamics with liquid time-constants. 100x faster than neural ODEs while matching or beating RNN/LSTM/GRU. [Paper](https://www.nature.com/articles/s42256-022-00556-7)
- *Module connection:* Biological dynamics are continuous; neither discrete RNNs nor AIF's discrete-time formulation captures this. CfCs on cognitive tasks remain unexplored.

### Tiny Interpretable RNNs
- **Ji-An, Benna & Mattar**, *Nature* 2025. RNNs with 1–4 units outperform 30+ classical RL/Bayesian models on reward-learning tasks (reversal learning, drifting bandits, two-stage tasks). Full dynamical systems analysis: phase portraits, fixed points, limit cycles. [Paper](https://www.nature.com/articles/s41586-024-08145-x)
- **Katahira** (bioRxiv Oct 2025): Rebuttal arguing that tiny RNN flexibility creates interpretability risks.
- *Module connection:* Fixed points correspond to stable beliefs; the dynamics reveal strategies invisible to both classical RL and AIF model-fitting.

### Transformers as Cognitive Models
- **Whittington et al.** (arXiv 2024). Attention-only transformer on a working memory task spontaneously learns input gating (key composition) and output gating (query composition) mirroring frontostriatal circuits. 70.2% attention on the relevant stored register.
- *Module connection:* Gating mirrors the A-matrix's role in AIF — filtering what enters and exits the belief state (Module 8).

### Multiplicative Couplings
- bioRxiv July 2025. Thalamocortical-inspired multiplicative interactions between feedforward and recurrent pathways. Hebbian-weight amplification for context-dependent gating. Dramatic improvements in learning speed on working memory and attention tasks.

### Symbolic Mechanisms in LLMs
- **Yang, Campbell, Webb & Cohen**, *ICML* 2025. Three-stage symbolic architecture emerges in trained LLMs: symbol abstraction heads → symbolic induction heads → retrieval heads. [arXiv](https://arxiv.org/abs/2502.01235)
- *Module connection:* Resolution to the symbolic vs. connectionist debate — structured computation emerges from optimization.

### Continuous Thought Machines
- **Sakana AI**, May 2025. Each neuron processes temporal signal history; information encoded in synchronization patterns. Adaptive compute time (harder inputs = more internal ticks), analogous to human reaction-time effects.

### Dynamical Similarity Analysis (DSA)
- **Ostrow et al.**, *NeurIPS* 2023. Koopman/DMD spectral methods compare trained networks at the level of dynamics, not geometry. DSA is the only method that distinguishes different dynamic implementations of the same computation (CKA, Procrustes cannot). [arXiv](https://arxiv.org/abs/2301.10197)
- *Module connection:* Could compare RL vs. AIF agent internals at the dynamics level (Module 13 extension).

## 6. Open Research Directions

These gaps were identified in Module 7 and are directly addressable with the tools in this curriculum:

### Gap 1: Systematic Architecture Comparison on Cognitive Tasks
No one has trained RNNs, SSMs, transformers, and continuous-time networks on the *same* classical cognitive tasks and compared internal representations against neural recordings. **Tools needed:** NeuroGym (tasks), PyTorch/JAX (training), DSA (analysis).

### Gap 2: Continuous-Time Models on Standard Cognitive Tasks
CfCs and Liquid Time-Constant networks have not been applied to the standard cognitive task battery. Biological dynamics are continuous, yet all current cognitive modeling uses discrete-time networks. **Tools needed:** CfC implementation, NeuroGym tasks.

### Gap 3: Differentiable AIF Agents on the NeuroGym Battery
Deep Active Inference (Module 12) has not been tested on neuroscience-standard tasks. Running AIF agents on PerceptualDecisionMaking, DelayMatchSample, and ContextDecisionMaking would be a landmark result. **Tools needed:** `alf` framework (Module 12), NeuroGym tasks.

### Gap 4: Koopman Spectral Analysis on Cognitive Networks
DSA (Ostrow et al.) has been applied to simple tasks (Flip-Flop). Applying it to networks trained on parametric working memory or context-dependent integration could reveal how different architectures implement the *same* cognitive computation through different dynamics. **Tools needed:** DSA implementation, trained networks from Gap 1.
